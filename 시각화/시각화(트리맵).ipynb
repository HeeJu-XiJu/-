```python
df=pd.read_csv('/content/drive/MyDrive/뽀로로드/data(아파트시세_가공)/final.csv', encoding='cp949')
df.head()
```

```python
df_tree1 = df[['동','평당전세가','자치구명']]
df_tree1
```

```python
df_tree2 = df_tree1.groupby(['자치구명','동']).mean().reset_index()
print(df_tree2)
```

```python
# import plotly.express as px
# https://plotly.com/python/getting-started/

fig = px.treemap(df_tree2,
                 path=['자치구명','동'],#위계구조 순서별 칼럼
                 values='평당전세가', # 사각형의 크기
                 color='평당전세가', # 사각형의 색상의 농도
                 color_continuous_scale='viridis')

fig.show()
```

```python
df_tree01 = df[['자치구명','동','연식','평당전세가']]
df_tree02 = df_tree01.groupby(['자치구명']).mean().reset_index()
print(df_tree02)
```

```python
fig = px.treemap(df_tree02,
                 path=['자치구명'],
                 values='평당전세가', # 사각형의 크기
                 color='연식', # 사각형의 색상의 농도
                 color_continuous_scale='Blues',
                 title='구별 평당전세가와 연식의 관계') # 기존색상 Viridis

fig.show()
```

```python
df5 = df_tree01[df_tree01['자치구명'].isin(['강남구', '강북구', '강동구','관악구', '강서구'])]
df5
```

```python
fig = px.treemap(df5,
                 path=['자치구명','동'],
                 values='평당전세가', # 사각형의 크기
                 color='연식', # 사각형의 색상의 농도
                 color_continuous_scale='Blues',
                 title='구별 평당전세가와 연식의 관계 TOP5') # 기존색상 Viridis

fig.show()
```
