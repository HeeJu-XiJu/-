```python
# 데이터 예시 (전체 합이 100% 로 나와야 함)
data_percentages = {'연식': 39, '버스정류장': 20, '스타벅스': 13, '어린이집': 13, '계약일': 12}

# Waffle 차트 생성
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data_percentages,
    colors=["#232066", "#983D3D", "#DCB732", "#7D7F75", "#8dd3c7"],
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1), 'fontsize': 10},
    icons='star',
    font_size=28, # 아이콘 크기
    title={'label': 'Feature Importance TOP5', 'loc': 'center'},
    labels=["{} ({:.1f}%)".format(k, v) for k, v in data_percentages.items()],
    figsize=(10, 10)

)

plt.show()
```
