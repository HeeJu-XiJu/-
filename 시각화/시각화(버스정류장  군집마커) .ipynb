```python
# 서울 지도에 버스정류장 지점 수 시각화
# MarkerCluster
map = folium.Map(location=[37.468656, 126.886638], zoom_start=12, width='%100', height='%100')
locations = list(zip(df.Y좌표, df.X좌표)) # 위, 경도 열을 zip으로 묶어서 list화 시킴
# from folium import plugins
cluster = plugins.MarkerCluster(locations=locations,
                                popups=df["정류소명"].tolist()) # 각 구역에 존재하는 버스정류장 지점의 수를 숫자로 표현
map.add_child(cluster)
# 지도를 더 축소하면 원들이 뭉쳐 표현되면서 숫자들이 합해짐 / 지도를 확대하면 원이 쪼개지다가 결국 하나의 핀으로 표현
map
```
